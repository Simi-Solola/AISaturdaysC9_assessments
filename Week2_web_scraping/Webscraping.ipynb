```python
import requests

# Send an HTTP GET request
url = "https://en.wikipedia.org/wiki/List_of_universities_in_Nigeria"
response = requests.get(url)

# Check the response status code
print("Response Status Code:", response.status_code)

# Print the response content
print("Response Content:")
print(response.text)
```

```python
from bs4 import BeautifulSoup

# Parse the HTML content
soup = BeautifulSoup(response.content, "html.parser")

# Print the prettified HTML
print(soup.prettify())
```

```python
table_1 = soup.find_all("table")[0]
```

```python
header_1 = table_1.find_all("th")
header_1
```

```python
header_1_data = [title.text.strip() for title in header_1]
```

```python
header_1_data
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns=header_1_data)
```
